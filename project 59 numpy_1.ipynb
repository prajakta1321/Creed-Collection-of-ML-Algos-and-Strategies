```python
import numpy as np
import pandas as pd

# Load the dataset
data = pd.read_csv('BHP.csv')
data.head()
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
data.head()
```

```python
# Drop rows with NaN values
# Fill NaN values with mean of each column
data.fillna(data.mean(), inplace=True)
data
```

```python
X = data[['total_sqft','bath','balcony']]
y = data[['price']]
```

```python
X_scaled = (X - np.mean(X, axis=0)) / np.std(X, axis=0) # scaling the features
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# centering the data and ensuring that the features are on a comparable level

X_train_mean = np.mean(X_train, axis=0)
X_train_centered = X_train - X_train_mean
```

```python
theta = np.dot(np.dot(np.linalg.inv(np.dot(X_train_centered.T, X_train_centered)), X_train_centered.T), y_train)
# this line of code computes the parameter vector theta using the OLS method(ordinal least squares)
```

```python
X_test_centered = X_test - X_train_mean
y_pred = np.dot(X_test_centered, theta)
y_pred
```

```python
X_test_centered
```

```python
nan_count = np.count_nonzero(np.isnan(y_pred))
print(f"Number of NaN values in y_pred: {nan_count}")
```

```python
y_pred[np.isnan(y_pred)] = np.nanmean(y_pred)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
