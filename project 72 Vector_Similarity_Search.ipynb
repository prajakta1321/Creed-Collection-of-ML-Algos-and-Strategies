```python

import matplotlib.pyplot as plt
import numpy as np
```

Vector similarity search is a method used to find similar items in a dataset by comparing vectors, which are mathematical representations of data.

```python
from mpl_toolkits.mplot3d import Axes3D
```

 Vector Representation: Each item in the dataset is represented as a vector.

```python
# Example pets attributes: [weight, height, age]

sheep = [9, 19, 3]
goat = [6, 12, 5]
```

# Similarity Measure:

To find items that are similar to a given query item, a similarity measure is used.

Common measures include:

# Cosine Similarity:

Measures the cosine of the angle between two vectors, indicating how similar they are in terms of direction.

# Euclidean Distance:

Measures the straight-line distance between two points in a vector space.

# Manhattan Distance:

Measures the sum of the absolute differences of their coordinates.

```python
# making a 3d plot

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(sheep[0], sheep[1], sheep[2], label="sheep", c='teal')
ax.scatter(goat[0], goat[1], goat[2], label="goat", c='gold')


ax.quiver(0, 0, 0, sheep[0], sheep[1], sheep[2], color='teal',
arrow_length_ratio=0.1)

ax.quiver(0, 0, 0, goat[0],  goat[1], goat[2], color='gold',
arrow_length_ratio=0.1)


ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Height (cm)')
ax.set_zlabel('Age (years)')


ax.set_xlim(0, 10)
ax.set_ylim(0, 40)
ax.set_zlim(0, 5)


ax.legend()
ax.set_title('3D Representation of Pets')

plt.show()
```
